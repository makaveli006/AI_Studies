```python
# import the Libraries
import numpy as np                # used for vMulti dimensional array
import pandas as pd               # used for import the dataset
import matplotlib.pyplot as plt   # used for ploting the Graph
from sklearn import metrics
```

```python
dataset=pd.read_csv('second_hand_Car_price.csv')
```

```python
dataset.head()
```

```python
ds=dataset.drop(['v.id'],axis=1)
```

```python
ds.head()
```

```python
x=ds.iloc[:,:-1].values   
y=ds.iloc[:,-1].values 
```

```python
x
```

```python
y
```

```python
# spliting the dataset into test set and train set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
regressor
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred_test=regressor.predict(x_test)
```

```python
y_pred_train=regressor.predict(x_train)
```

```python
y_pred_train
```

```python
y_train
```

```python
y_train
```

```python
y_pred_train
```

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_train,y_pred_train))
```

```python
print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_pred_test))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred_test)))
```
