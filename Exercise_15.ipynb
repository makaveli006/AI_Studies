```python

pd['Self_Employed'].fillna(pd['Self_Employed'].mean(),inplace=True)
pd['LoanAmount'].fillna(pd['LoanAmount'].mean(),inplace=True)
pd['Loan_Amount_Term'].fillna(pd['Loan_Amount_Term'].mean(),inplace=True)
pd['Credit_History'].fillna(pd['Credit_History'].mean(),inplace=True)
```

```python
pd.isnull().sum()
```

```python
x = pd.iloc[:,0:-1].values
y = pd.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
x.shape, y.shape
```

```python
model = Sequential()
model.add(Dense(12, input_shape=(10,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model
```

```python
model.fit(x, y, epochs=50, batch_size=10)
```

```python
_, accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions = (model.predict(x) > 0.5).astype(int)
```
