```python
# Import necessary Kivy modules
from kivy.app import App
from kivy.uix.button import Button
from kivy.uix.label import Label
from kivy.uix.boxlayout import BoxLayout

# Define the main app class
class SimpleKivyApp(App):
    def build(self):
        # Create a layout (in this case, a BoxLayout)
        layout = BoxLayout(orientation='vertical', padding=10)

        # Create a label
        self.label = Label(text="Hello, Kivy!")

        # Create a button
        btn = Button(text="Click me!")
        # Bind the button's on_press event to a callback function
        btn.bind(on_press=self.on_button_click)

        # Add widgets to the layout
        layout.add_widget(self.label)
        layout.add_widget(btn)

        # Return the layout as the root widget of the application
        return layout

    # Callback function for the button click event
    def on_button_click(self, instance):
        self.label.text = "Button clicked!"

# Run the app
if __name__ == '__main__':
    SimpleKivyApp().run() vaa
```
